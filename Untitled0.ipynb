```python
X = data_frame.iloc[:, 2:31].values 
Y = data_frame.iloc[:, 1].values 
```

```python
# Split the data again, but this time into 75% training and 25% testing data sets.
# 75% Training Set and 25% Test Setd

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
def models(X_train, Y_train):
# Using SVC method of svm class to use Support Vector Machine Algorithm
  from sklearn.svm import SVC
  from xgboost import XGBClassifier
  from sklearn.metrics import confusion_matrix, accuracy_score
  from sklvq import GLVQ

  def get_acc(Y_test, Y_pred):
    acc = accuracy_score(Y_test, Y_pred) * 100
    acc = round(acc, 2)
    return acc

  classifier = SVC(kernel = 'linear', random_state = 0).fit(X_train, Y_train)
  Y_pred = classifier.predict(X_test)
  cm = confusion_matrix(Y_test, Y_pred)
  acc = get_acc(Y_test, Y_pred)

  classifier_xgb = XGBClassifier().fit(X_train, Y_train)
  Y_pred_xgb = classifier_xgb.predict(X_test)
  cm_xgb = confusion_matrix(Y_test, Y_pred_xgb)
  acc_xgb = get_acc(Y_test, Y_pred_xgb)

  classifier_LVQ = GLVQ().fit(X_train, Y_train)
  Y_pred_lvq = classifier_LVQ.predict(X_test)
  cm_lvq = confusion_matrix(Y_test, Y_pred_lvq)
  acc_lvq = get_acc(Y_test, Y_pred_lvq)

  
  def get_output():
    print("SVM Accuracy: {}".format(acc))
    print("LVQ Accuracy: {}".format(acc_lvq))
    print("XGBoost Accuracy: {}".format(acc_xgb))
  

  return get_output()
```

```python
model = models(X_train,Y_train)
model
```
